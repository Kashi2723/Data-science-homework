```python
# 1. What is the role of try and exception block?

Ans. The try block lets us test a block of code for errors. Whereas, The except block lets us handle the error.
```

```python
# 2. What is the syntax for a basic try-except block?

Ans. 

try:
    # some code that may raise an exception
except ExceptionType:
    # code to handle the exception
```

```python
# 3. What happens if an exception occurs inside a try block and there is no matching except block?

Ans. If an exception occurs inside a try block and there is no matching except block, the program will terminate with an unhandled exception error. 
The rest of the code in the try block will be skipped, and any finally block will be executed before the program stops. 
```

```python
# 4. What is the difference between using a bare except block and specifying a specific exception type?

Ans. Using a bare except block (except:) and specifying a specific exception type (except SomeException:) serve different purposes in handling exceptions.

As bare except block catches any type of exception occurs within the corresponding 'try' block. It is a general catch-all, and will handle all exceptions regardless of their type.
Ex :
        try:
            # some code that may raise an exception
        except:
            # handle any exception
        
As in specifying a specific exception type (except SomeException: ), catches only the specified type of exception (e.g., except ValueError:). 
It allows us to handle different exceptions differently, providing more control over the exception-handling process.
Ex : 
        try:
            # some code that may raise a ValueError
        except ValueError:
            # handle only ValueError


```

```python
# 5. Can you have nested try-except blocks in Python? If yes, then give an example.

Ans. 

try:
    # Outer try block
    num = input("Enter a number: ")

    try:
        # Inner try block
        num1 = int(num)
        result = 10 / num1
        print("Result:", result)

    except ZeroDivisionError:
        print("Cannot divide by zero in the inner try block.")

except ValueError:
    print("Invalid input. Please enter a valid number.")
```

```python
# 6. Can we use multiple exception blocks, if yes then give an example.

Ans. 

try:
    num1 = int(input("Enter your num1: "))
    num2 = int(input("Enter your num2: "))
    res = num1/num2
    print("Result:",res)

except ValueError:
    print("Invalid input. Enter only numbers")
except ZeroDivisionError:
    print("Cannot divide by Zero")
except Exception as e:
    print("An error occured,"e)
```

```python
# 7. Write the reason due to which following errors are raised:

(a) EOFError : EOFError is a type of error which is raised when an operation that expects the end of a file is performed, 
but the end of the file is not encountered as expected.

(b) FloatingPointError : FloatingPointError is a type of error is  raised in Python when a floating-point operation fails to produce a valid result. 
This typically occurs in situations where the result of a floating-point calculation is undefined or not representable due to limitations in the floating-point representation.

(c) IndexError : IndexError is a type of error is raised when you try to access an index in a sequence (such as a list, tuple, or string) that is outside the valid range of indices.

(d) MemoryError : MemoryError is a type of error is raised when an operation or expression attempts to allocate more memory than the system can provide. 
This typically occurs when your program exceeds the available memory resources, and the Python interpreter cannot fulfill a request for additional memory.

(e) OverflowError : OverflowError is a type of error is raised when the result of an arithmetic operation exceeds the representational limits of the data type involved. 
This typically occurs with integer operations that go beyond the maximum or minimum values that can be stored.

(f) TabError : TabError is a type of error is raised when there is an issue related to the use of tabs and spaces for indentation. Python relies on consistent indentation 
to define the structure of the code, and mixing tabs and spaces in an inconsistent manner can lead to this error.

(g) ValueError : ValueError is a type of error is raised when an operation or function receives an argument of the correct type but with an invalid or inappropriate value. 
This exception is often used to indicate that a function cannot work with the provided input value because it is outside the expected range or doesn't meet certain criteria.
```
